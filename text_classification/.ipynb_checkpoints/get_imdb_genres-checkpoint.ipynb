Grab movie genres for classification exercise. Genres are on the imdb web page so we just need to pull them off based on urls provided in the dataset.

```python
# Standard imports needed 
# for this exercise
import numpy as np
import pandas as pd

from tqdm import tqdm
from collections import Counter

import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib_venn import venn2

import requests
import re
from bs4 import BeautifulSoup
import ast
```

### Read imdb urls into memory

```python
# need to extract name of movie from data
directory = "~/git_base/python/text_classification/stanford_data/aclImdb/train/"
fname = "urls_pos.txt"
df_pos_urls = pd.read_csv(directory + fname, encoding='utf-8', header=None)
df_pos_urls.columns = ['url']
df_pos_urls.head()
```

### Function to scrape movie genres from imdb.com

```python
def scrape_movie_genres(df):
    """
    use url provided in df
    to scrape movie genre 
    from imdb page
    """

    # initialize the array
    movie_data = []

    # loop over each movie
    for row in tqdm(df.iterrows(), total = len(df)):

        # request the html data
        # need to remove excess url stuff
        response = requests.get(row[1]['url'].split('/usercomments')[0])

        # parse the text
        soup = BeautifulSoup(response.text, 'html.parser')

        # genre data is in the script tag
        scriptrows = soup.find_all('script')
        try:
            # look in the first one
            cand = [r for r in scriptrows if str(r).split('genre')[0] != str(r)][0]    
            try:
                # multiple genres
                genres = ast.literal_eval(str(cand).split('genre')[1].split(']')[0].replace("\n","").replace('": [',"").lstrip().rstrip())
            except SyntaxError:
                # hit an error
                # because only one genre
                genres = (str(cand).split('genre')[1].split(',')[0].replace("\n","").replace('":',"").lstrip().rstrip().replace('"',""))

            # grab the movie title
            movie_title = soup.title.text.replace(' - IMDb', "")

            # record the data
            movie_data.append((row[0], movie_title, genres))
        except:
            # no genres found -- skip this entry
            print "Error with %i" % row[0]

    return movie_data
```

```python
pos_movie_data = scrape_movie_genres(df_pos_urls)
df_pos_genres = pd.DataFrame(data=pos_movie_data,
                             columns = (('ref', 'title', 'genres')))
df_pos_genres.head()
```

```python
# quick clean up
clean_data = []
for row in df_pos_genres.iterrows():
    if type(row[1]['genres']) == tuple:
        clean_data.append(row[1]['genres'])
    else:
        clean_data.append([row[1]['genres']])
        
df_pos_genres['genres2'] = clean_data
df_pos_genres['genres'] = clean_data
df_pos_genres = df_pos_genres.drop('genres2', axis=1)
df_pos_genres.head()
```

### One-Hot Encode movie genres

```python
"""
create indicator column for each genre
"""

# flatten all genres
flat_genres = [item for sublist in df_pos_genres['genres'].values for item in sublist]

# grab unique ones
# plus a cleanup step
unique_genres = np.unique(flat_genres)
unique_genres = [u for u in unique_genres if 'could speak for itself' not in u]

def indicate_genre(col, genre):
    if genre in col:
        return 1
    else:
        return 0

for genre in tqdm(unique_genres, total = len(unique_genres)):
    df_pos_genres[genre] = df_pos_genres['genres'].apply(indicate_genre, genre = genre)
```

```python
df_pos_genres.to_csv('stanford_train_pos_genres.csv', encoding='utf-8')
df_pos_genres.head()
```

```python
Counter([item for sublist in clean_data for item in sublist]).most_common()
```

### Visualizations

```python
df_pos_genres_unique = df_pos_genres.drop_duplicates('title')
counts_data = Counter([item for sublist in df_pos_genres_unique['genres'] for item in sublist]).most_common()
counts_data = [cd for cd in counts_data if cd[1] > 10]
vals = [cd[1] for cd in counts_data]

plt.figure(figsize=(12, 9))
ax = plt.subplot(111)  
ax.spines["top"].set_visible(False)  
ax.spines["right"].set_visible(False)

ax.get_xaxis().tick_bottom()  
ax.get_yaxis().tick_left() 

plt.xticks(range(0, 30, 1), [cd[0] for cd in counts_data], fontsize=14, rotation = 70)  
plt.yticks(fontsize=14)

plt.xlabel("Movie Genre", fontsize=16)  
plt.ylabel("Count of Occurances", fontsize=16)
plt.title('IMDB Genres for Stanford Data Set (Positive Training Set)', fontsize=16)

plt.bar(np.arange(len(counts_data)), vals, width=0.9, facecolor = "#3F5D7D" )
```

```python
cooc_vals = []
for genre1 in unique_genres:
    row_vals = []
    for genre2 in unique_genres:
        g1rows = df_pos_genres_unique.loc[
                    df_pos_genres_unique[genre1] == 1]
        g2rows = g1rows.loc[g1rows[genre2] == 1]
        row_vals.append(1.0 * len(g2rows) / len(g1rows))
    cooc_vals.append(row_vals)
df_cooc = pd.DataFrame(data=cooc_vals, index=unique_genres, columns=unique_genres)    
sns.clustermap(data=df_cooc, cmap='Blues', metric = 'minkowski')
```
